# The Data:

In this example we use the MNist dataset from torchvision. These are Pil Images, therefore we need to do the datapreprocessing of turning them into torchtensors.

the data consists of torch tensors

```python
transform = transforms.Compose([
    # you can add other transformations in this list
    transforms.ToTensor()
])
imagenet_data = torchvision.datasets.MNIST('../data', download=True, transform=transform)
data_loader = torch.utils.data.DataLoader(imagenet_data,
                                          batch_size=1,
                                          shuffle=True,
                                          num_workers=0) 
# num workers depends on your cpu, dont set it too high (more memory usage) or too low (gpu idling) it will slow down training 
# each worker loads a single batch at once
# num workers = 0, main process does data loading when needed
print(len(data_loader))
```

Let's plot an image:

```python
x_sample, y_sample = next(iter(data_loader))
plt.imshow(x_sample[0].reshape((28, 28)), cmap="gray")
print(x_sample.shape)
```
